```python
# Identifying outliers in enrollment numbers for each study status
# Using the IQR method to find outliers
outliers = {}
for status in data['Status'].unique():
    subset = data[data['Status'] == status]['Enrollment'].dropna()
    Q1 = subset.quantile(0.25)
    Q3 = subset.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers[status] = subset[(subset < lower_bound) | (subset > upper_bound)]

# Displaying the outliers for each status
for status, outlier_values in outliers.items():
    print("Status:", status)
    print("Outliers:", outlier_values.values)
    print("\
")
```

```python
# Function to remove outliers using IQR method
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

# Create a copy of the original dataframe
data_no_outliers = data.copy()

# Remove outliers for each status group
status_groups = []
for status in data_no_outliers['Status'].unique():
    status_data = data_no_outliers[data_no_outliers['Status'] == status]
    clean_status_data = remove_outliers(status_data, 'Enrollment')
    status_groups.append(clean_status_data)

# Combine all clean status groups
data_no_outliers = pd.concat(status_groups)

# Create box plot without outliers
plt.figure(figsize=(15, 8))
sns.boxplot(x='Status', y='Enrollment', data=data_no_outliers)
plt.xticks(rotation=45, ha='right')
plt.title('Distribution of Enrollment Numbers by Study Status (Outliers Removed)')
plt.xlabel('Study Status')
plt.ylabel('Enrollment Numbers')
plt.tight_layout()
plt.show()

# Print summary statistics of the cleaned data
print("\
Summary statistics after removing outliers:")
print(data_no_outliers.groupby('Status')['Enrollment'].describe())
```

```python
# Data wrangling operations
import pandas as pd
import numpy as np

# Load the data
df = pd.read_csv('/content/drive/MyDrive/COVID clinical trials.csv')

# 1. Convert date columns to datetime
date_columns = ['Start Date', 'Primary Completion Date', 'Completion Date',
                'First Posted', 'Results First Posted', 'Last Update Posted']

for col in date_columns:
    df[col] = pd.to_datetime(df[col], errors='coerce')

# 2. Clean Enrollment column (should be numeric)
df['Enrollment'] = pd.to_numeric(df['Enrollment'], errors='coerce')

# 3. Create a simplified status category
df['Status_Category'] = df['Status'].map({
    'Completed': 'Completed',
    'Recruiting': 'Active',
    'Not yet recruiting': 'Active',
    'Active, not recruiting': 'Active',
    'Withdrawn': 'Terminated',
    'Terminated': 'Terminated',
    'Suspended': 'Terminated'
}).fillna('Other')

# 4. Clean phases data
df['Phases'] = df['Phases'].fillna('Not Specified')

# 5. Create study duration in days
df['Study_Duration'] = (df['Completion Date'] - df['Start Date']).dt.days

# Display basic information about the cleaned dataset
print("\
Dataset Info:")
print(df.info())

print("\
Sample of cleaned data:")
print(df[['NCT Number', 'Status_Category', 'Enrollment',
          'Start Date', 'Study_Duration', 'Phases']].head())
```

```python
# Chart - 17 Correlation Heatmap

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load your dataset (replace 'your_file.csv' with your actual file path)
df = pd.read_csv('/content/drive/MyDrive/COVID clinical trials.csv')

# Select numerical columns for correlation analysis (replace with your relevant columns)
numerical_cols = df.select_dtypes(include=['number']).columns

# Calculate the correlation matrix
correlation_matrix = df[numerical_cols].corr()

# Create the heatmap
plt.figure(figsize=(12, 10))  # Adjust figure size as needed
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Heatmap')
plt.show()
```

**1. Why did you pick the specific chart?**

 I chose a correlation heatmap to visualize the relationships between numerical variables in the clinical trial dataset.  A heatmap is an excellent way to represent correlation coefficients between multiple variables simultaneously.  The color intensity and the numerical annotations on the heatmap clearly show the strength and direction of the correlations, making it easy to identify strongly correlated variables and potential multicollinearity issues.

**2. What is/are the insight(s) found from the chart?**

  **The insights from the correlation heatmap:**

  * **Strength of Correlation:** The color intensity of each cell in the heatmap indicates the strength of the correlation between two variables.  Darker colors (e.g., dark red or dark blue) represent stronger correlations, while lighter colors represent weaker correlations.

  * **Direction of Correlation:**  The color scale (e.g., coolwarm) indicates the direction of the correlation.  Positive correlations (variables tend to increase or decrease together) are typically shown in warm colors (reds), while negative correlations (one variable increases as the other decreases) are shown in cool colors (blues).  Values close to zero suggest little to no linear relationship between the variables.

  * **Multicollinearity:**  The heatmap makes it easy to identify pairs of highly correlated variables, which could indicate multicollinearity. Multicollinearity is a statistical phenomenon in which multiple independent variables in a regression model are highly correlated.  This can make it difficult to isolate the individual effect of each variable on the dependent variable.

  * **Relationships between Numerical Variables:**  The heatmap provides a comprehensive overview of the relationships between all pairs of numerical variables, allowing for quick identification of significant correlations and potential relationships that might warrant further investigation.
